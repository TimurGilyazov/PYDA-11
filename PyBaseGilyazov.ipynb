**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

```python
month = input ('Введите месяц Вашего рождения с маленькой буквы: ')
date = int (input ('Введите число месяца: '))
if (month == 'март' and date >= 21) or (month == 'апрель' and date <=20):
    print ('Ваш знак Зодиака: Овен')
elif (month == 'апрель' and date >= 21) or (month == 'май' and date <=21):
    print ('Ваш знак Зодиака: Телец')
elif (month == 'май' and date >= 22) or (month == 'июнь' and date <=21):
    print ('Ваш знак Зодиака: Близнецы')
elif (month == 'июнь' and date >= 22) or (month == 'июль' and date <=23):
    print ('Ваш знак Зодиака: Рак')
elif (month == 'июль' and date >= 24) or (month == 'август' and date <=23):
    print ('Ваш знак Зодиака: Лев')
elif (month == 'август' and date >= 24) or (month == 'сентябрь' and date <=23):
    print ('Ваш знак Зодиака: Дева')
elif (month == 'сентябрь' and date >= 24) or (month == 'октябрь' and date <=23):
    print ('Ваш знак Зодиака: Весы')
elif (month == 'октябрь' and date >= 24) or (month == 'ноябрь' and date <=22):
    print ('Ваш знак Зодиака: Скорпион')
elif (month == 'ноябрь' and date >= 23) or (month == 'декабрь' and date <=21):
    print ('Ваш знак Зодиака: Стрелец')
elif (month == 'декабрь' and date >= 22) or (month == 'январь' and date <=20):
    print ('Ваш знак Зодиака: Козерог')
elif (month == 'январь' and date >= 21) or (month == 'февраль' and date <=19):
    print ('Ваш знак Зодиака: Водолей')
elif (month == 'февраль' and date >= 20) or (month == 'март' and date <=20):
    print ('Ваш знак Зодиака: Рыбы')
else:
    print ('Вы указали некорректный месяц или число')
```

**Задание 4**. Нужно разработать приложение для финансового планирования.
Приложение учитывает сколько уходит на ипотеку, "на жизнь" и сколько нужно отложить на пенсию.
Пользователь вводит:
- заработанную плату в месяц.
- сколько процентов от ЗП уходит на ипотеку.
- сколько процентов от ЗП уходит "на жизнь".
- сколько раз приходит премия в год.

Остальная часть заработанной платы откладывается на пенсию.

Также пользователю приходит премия в размере зарплаты, от которой половина уходит на отпуск, а вторая половина откладывается.

Программа должна учитывать сколько премий было в год.

Нужно вывести сколько денег тратит пользователь на ипотеку и сколько он накопит за год.

Пример:

Введите заработанную плату в месяц: 100000

Введите сколько процентов уходит на ипотеку: 30

Введите сколько процентов уходит на жизнь: 50

Введите количество премий за год: 2

Вывод:
На ипотеку было потрачено: 360000 рублей
Было накоплено: 340000 рублей

```python
print ('Добрый день! Вас приветствует Планировщик личных накоплений.')
salary = int (input ('Ведите размер ежемесячной заработной платы в рублях: ')) 
bonus = int (input ('Введите сколько раз в год Вы получаете премию в размере одной месячной зарплаты: '))
regular = int(input('Какую долю ежемесячной зарплаты Вы тратите "на жизнь" в процентах: '))
ipoteka = int(input('Какую долю ежемесячной зарплаты Вы тратите на ипотеку в процентах: '))
print ('За год Вы сможете отложить на пенсию: ', salary * 12 * ((100 - regular - ipoteka)/100) + salary * bonus * 0.5 , 
       'За год вы потратите на ипотеку: ',  salary * 12 * ipoteka / 100)
```
